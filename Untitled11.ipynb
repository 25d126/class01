```python

import matplotlib.pyplot as plt
import time
from IPython import display

def draw_square(robot_position, size=1):
    """Draws a square centered around the robot's position."""
    x, y = robot_position
    # Define the four points of the square
    square_x = [x - size/2, x + size/2, x + size/2, x - size/2, x - size/2]
    square_y = [y - size/2, y - size/2, y + size/2, y + size/2, y - size/2]
    return square_x, square_y

# Initialize robot position
robot_x, robot_y = 0, 0
step_size = 1

# Create the plot
plt.figure(figsize=(6, 6))
ax = plt.gca()
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_aspect('equal', adjustable='box')
ax.set_title("Moving Robot Vacuum Drawing a Square")

# Simulate the robot movement and drawing
path_x = [robot_x]
path_y = [robot_y]

# Move right
for _ in range(1):
    robot_x += step_size
    path_x.append(robot_x)
    path_y.append(robot_y)
    square_x, square_y = draw_square((robot_x, robot_y))
    plt.plot(square_x, square_y, 'b-', alpha=0.2) # Draw the square boundary
    plt.plot(path_x, path_y, 'ro-') # Draw the robot's path
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.5)

# Move up
for _ in range(1):
    robot_y += step_size
    path_x.append(robot_x)
    path_y.append(robot_y)
    square_x, square_y = draw_square((robot_x, robot_y))
    plt.plot(square_x, square_y, 'b-', alpha=0.2) # Draw the square boundary
    plt.plot(path_x, path_y, 'ro-') # Draw the robot's path
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.5)
```
